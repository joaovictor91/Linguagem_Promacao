### 1. Elabore uma estrutura para representar e armazenar 10 produtos (código, nome, preço). Utilize os recursos de arquivo para armazenar estes dados permanentemente. O nome do arquivo deve ser o mesmo da estrutura. 

```
class produto:
    cod=0
    nome=''
    preco=0.0

def main():
    total=10
    vet=[]
    pro=open('produtos.txt','w')
    
    
    print("Cadastros de Produtos")
    for i in range(total):
        p=produto()
        p.cod=int(input("Informe o código do produto: "))
        p.nome=input("Informe o nome do produto: ")
        p.preco=float(input("Informe o preço do produto: R$"))
        pro.write('%d,%s,%.2f\n' % (p.cod, p.nome, p.preco))
        vet.append(p)
        
    pro.close()

    
    pro = open('produtos.txt', 'r')

    print('Apresentaçao dos dados dos Produtos.........')



    print('Código\t\tNome\t\tPreço')

    for linha in pro.readlines():

        cod, nome, preco = linha.strip().split(',') # linha = [6 Lia 100]

        print(cod,'\t\t',nome,'\t\tR$',preco)         # linha 45 JosÉ 4000.00

        #print(linha)

    pro.close()

main()
```

### 4.Baixe o arquivo futebol.txt que esta na pasta Material do aula do Teams. Leia este arquivo e o apresente. Apresente este novo arquivo.

```
class jog:
    posicao_jogo=''
    altura=0.0
    peso=0.0
    imc=0.0
    
def Exibir_time():
    futebol=open('futebol.txt','r')
    print('Apresentaçao dos dados dos Jogadores de Futebol.........')
    vet=[]
    print('Posição do Jogador\tAltura\t\tPeso')
    futebol_imc=open('futebol_imc.txt','w')
    for linha in futebol.readlines():
        
        
        posicao_jogo,altura ,peso = linha.strip().split(',') 
        
        print(posicao_jogo,'\t\t',altura,'\t\t',peso)         # linha 45 JosÉ 4000.00
        j=jog()
        j.posicao_jogo=posicao_jogo
        j.altura=altura
        j.peso=peso
        imc=float(j.peso)/(float(j.altura)*float(j.altura))
        j.imc=str(imc)
        futebol_imc.write('%s,%.2f,%.2f,%.2f\n' % (j.posicao_jogo, float(j.altura), float(j.peso), float(j.imc)))
        vet.append(j)
        
    futebol_imc.close()    
    futebol.close()
    



def Exibir_imc():

    futebol_imc=open('futebol_imc.txt','r')
    print('Apresentaçao dos dados dos Jogadores de Futebol.........')



    print('Posição do Jogador\tAltura\t\tPeso\t\tImc')

    for linha in futebol_imc.readlines():

        posicao_jogo,altura ,peso, imc = linha.strip().split(',')  

        print(posicao_jogo,'\t\t',altura,'\t\t',peso,'\t\t',imc)         # linha 45 JosÉ 4000.00

        #print(linha)

    futebol_imc.close()



def main():

    op = 1

    while op !=3:

        print('\n\n*** Gerenciamento de Jogadores de Futebol COM ARQUIVO ***')

        print('1- Cadastrar')

        print('2- Mostrar dados do arquivo')

        print('3- Sair')

        op = int(input('Digite a opção: '))

        if op == 1:

            Exibir_time()

        elif op == 2:

            Exibir_imc()            
        elif op==3:
            print("Progama Finalizado")

        elif op>3:
            print("Opção Inválido")


main()
```
